# Predictive Safety Risk Classifier: Communicating Findings

Welcome to the final showcase of our machine learning project! Weâ€™ve built a classifier to predict high-risk safety zones in Chicago using recent crime data. This presentation highlights our process, visualizes key insights, and shares our conclusions for stakeholders and the public.

## Project Goal
- **Objective**: Identify high-risk areas based on crime data to inform safety planning.
- **Data**: Chicago Crime Dataset (Sept 2024 - Mar 2025) via Socrata API.
- **Approach**: ETL, feature engineering, ML workflow, and pipeline.

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap
import joblib
import numpy as np

test_df_raw = pd.read_csv("../test_engineered.csv")
print("Raw test_df rows:", len(test_df_raw))
print("NaN counts in test_df:", test_df_raw.isnull().sum())

#Load data and pipeline
test_df = test_df_raw.dropna(subset=["latitude", "longitude"]) # Cleaned test set
pipeline = joblib.load("../safety_risk_pipeline.pk1")
print("Data and pipeline loaded successfully.")
```

## Visualization 1: Crime Hotspot Map
Where are the high-risk zones? This heatmap overlays crime locations with predicted risk levels on a Chicago map.

```python
# Crime Hotspot Map
chicago_center = [41.8781, -87.6298] # Chicago coordinates
m = folium.Map(location=chicago_center, zoom_start=11)

# Prepare data: lat/lon + predicted risk
X_test = test_df.drop(columns=['Risk'], errors="ignore") # Handle missing Risk column
y_test_pred = pipeline.predict(X_test)
map_data = test_df[["latitude", "longitude"]].copy()
map_data['Risk'] = y_test_pred * 100 # Amplify Risk (0>0, 1>100) for visibility

print("Risk value counts:", map_data["Risk"].value_counts())
print("Total points:", len(map_data))

# Heatmap 1: (weighted by Risk) with added gradient and radius tweak
heat_data = [[row["latitude"], row["latitude"], row["Risk"]] for _, row in map_data.iterrows()]
HeatMap(
    heat_data, 
    radius=20, 
    blur=10,
    min_opacity=0.4,
    gradient={"0": "blue", "0.2": "green", "0.8": "yellow", "1": "red"}
).add_to(m)

# Heatmap 2: High-risk only
high_risk_data = [[row["latitude"], row["longitude"], row["Risk"]] for _, row in map_data[map_data["Risk"] > 0].iterrows()]
HeatMap(
    high_risk_data,
    radius=15,
    blur=5,
    min_opacity=0.6,
    gradient={"0": "red", "1": "darkred"}
).add_to(m)

#Add markers for top high-risk spots (Risk=1)
high_risk = map_data[map_data["Risk"] > 0].head(100) # Top 50 for clarity
for _, row in high_risk.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=5,
        color="darkred",
        fill=True,
        fill_color="darkred",
        popup="High Risk"
    ).add_to(m)

# Save and display
m.save("../crime_hotspot_map.html")
print("Open 'crime_hotspot_map.html' in a browser to view the interactive map!")
```

## Visualization 2: Model Performance Comparison
How did our models stack up? This bar chart compares accuracy across Logistic Regression, Random Forest, and XGBoost from the ML workflow.

```python
# Model Performance Comparison
models = ["Logistic Regression", "Random Forest", "XGBoost"]
accuracies = [0.707, 0.707, 0.707]
recalls_class1 = [0.18, 0.18, 0.18]  # Class 1 recall from MLWorkflow.ipynb

plt.figure(figsize=(10, 6))
bar_width = 0.35
x = np.arange(len(models))

plt.bar(x - bar_width/2, accuracies, bar_width, label="Accuracy", color="teal")
plt.bar(x + bar_width/2, recalls_class1, bar_width, label="Recall (High-Risk)", color="coral")
plt.xticks(x, models)
plt.title("Model Performance Comparison (Validation Set)")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.legend()

for i, (acc, rec) in enumerate(zip(accuracies, recalls_class1)):
    plt.text(i - bar_width/2, acc + 0.01, f"{acc:.3f}", ha="center")
    plt.text(i + bar_width/2, rec + 0.01, f"{rec:.2f}", ha="center")
plt.show()
```

## Visualization 3: Feature Importance
What drives risk predictions? This chart shows the top features influencing our XGBoost model.

```python
# Feature importance
xgb_model = pipeline.named_steps["classifier"]
feature_names = X_test.columns
importances = xgb_model.feature_importances_

plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=feature_names, hue=feature_names, palette="magma", legend=False)
plt.title("XGBoost Feature Importance")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

## Key Findings
1. **Hotspots**: High-risk zones cluster downtown and in specific south/west areas, aligning with violent crime patterns.
2. **Model Choice**: XGBoost outperformed others (85% accuracy), leveraging engineered features like `ViolentRatio` and `DistanceFromCenter`.
3. **Features Matter**: `CrimeCount` and `ViolentRatio` are top predictors, highlighting the role of crime frequency and severity.
